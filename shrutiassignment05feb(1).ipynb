```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

plt is the alias for pyplot.
numpy helps generate data for plotting.

```python
# Data
x = np.arange(0, 15, 1)  # X-axis: 0 to 9
y = x ** 2               # Y-axis: Square of x

# Basic Line Plot
plt.plot(x, y)
```

```python
# Data
x = np.arange(0, 10, 1)  # X-axis: 0 to 9
y = x ** 2               # Y-axis: Square of x

# Basic Line Plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis (Numbers)')
plt.ylabel('Y-axis (Squares)')
plt.title('Simple Line Plot for demonstration Example')

# Display the plot
plt.show()
```

```python
# Data for plotting a simple line chart
x = np.arange(0, 10, 1)  # X-axis: 0 to 9
y = x ** 2               # Y-axis: Square of x

# Basic Line Plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis (Numbers)')
plt.ylabel('Y-axis (Squares)')
plt.title('Simple Line Plot Example')

# Display the plot
plt.show()
# Key Functions:
# plot() â†’ Draws the line chart.
# xlabel(), ylabel() â†’ Labels for the axes.
# title() â†’ Adds a title.
# show() â†’ Displays the plot in Jupyter.
```

```python
# Custom Line Plot
x = np.arange(1, 11)  # Create an array of numbers from 1 to 10
y = x ** 2  # Create an array of the squares of x
plt.plot(x, y, color='blue', linestyle='--', marker='x')

# Adding labels and title
plt.xlabel('Numbers')
plt.ylabel('Squares')
plt.title('Customized Line Plot Example')

# Adding grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

```python
# Bar Chart: Participants from Different Cities
plt.figure(figsize=(10, 5))
plt.bar(df_participants['City'], df_participants['Participants'], color='red')
plt.xlabel('Cities')
plt.ylabel('Number of Participants')
plt.title('Participants from Different Cities - Khubh Mela')
plt.show()
```

```python
# Pie Chart: Distribution of Participants by City
plt.figure(figsize=(10, 8))
plt.pie(df_participants['Participants'], labels=df_participants['City'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Participant Distribution by City')
plt.show()
```

```python
# Multiple Plots in One Figure
plt.figure(figsize=(18, 9))

# Subplot 1: Line Plot
plt.subplot(1, 2, 1)
plt.plot(df_visitors['Day'], df_visitors['Visitors'], marker='s', color='red')
plt.title('Daily Visitors in kumbh')
plt.xlabel('Days')
plt.ylabel('Visitors')

# Subplot 2: Bar Chart
plt.subplot(1, 2, 2)
plt.bar(df_participants['City'], df_participants['Participants'], color='pink')
plt.title('City-wise Participants in kumbh')
plt.xlabel('Cities')
plt.ylabel('Participants')

plt.tight_layout()
plt.show()

# Adding Legends with Multiple Lines
plt.figure(figsize=(10, 5))
x = np.arange(1, 6)
plt.plot(x, visitors, label='Visitors Trend', linestyle='-', marker='o')
plt.plot(x, participants, label='Participants Trend', linestyle='--', marker='s')
plt.xlabel('Days/Cities Index')
plt.ylabel('Count')
plt.title('Visitors & Participants Trend - Khubh Mela')
plt.legend()
plt.grid(True)
plt.show()

# Saving a Plot
plt.figure(figsize=(10, 5))
plt.plot(df_visitors['Day'], df_visitors['Visitors'], color='purple')
plt.title('Saved Plot Example - Visitors Trend')
plt.savefig('khubh_mela_visitors_trend.png')
plt.show()
```
